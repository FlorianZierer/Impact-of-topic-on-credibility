```python
column1 = "1. Wie viel Vorwissen haben sie zu dem Thema des Textes?"
column2 = "2. Wie glaubwürdig fanden Sie den Text ?"
column3 = "3. Wie klar und verständlich war der Text?"
column4 = "4. Wie interessant fanden Sie den Text?"
column5 = "5. Wie viel Vertrauen haben Sie in den Inhalt des Textes?"
column6 = "6. Wie sachlich fanden sie den Text?"
column7 = "7. Haben Sie den Eindruck, dass der Autor Fachkompetenz besitzt?"
column8 = "8. Wie sehr vertrauen Sie den im Text präsentierten Informationen?"
column9 = "9. Wie wahrscheinlich ist es, dass der Text missverstanden werden könnte?"
column10 = "10. Begründen Sie ihre Entscheidungen. (optional)"

column_names = [column1,column2,column3,column4,column5,column6,column7,column8,column9,column10]

A = ["Sport","Geld","Wissen","Gesundheit"]
B = ["Geld","Gesundheit","Sport","Wissen"]
C = ["Gesundheit","Wissen","Geld","Sport"]
D = ["Wissen","Sport","Gesundheit","Geld"]

dataframes = [A1,B1,C1,D1,A2,B2,C2,D2]

```

```python
import re

def rename_columns(dataframe, column_names, label_list):
    new_column_names = []
    
    for column_name_df in dataframe.columns:
        matched = False
        for column_name in column_names:
            if column_name_df.startswith(column_name):
                # Extract the numeric suffix if present
                match = re.search(r'\.(\d+)$', column_name_df)
                number = int(match.group(1)) if match else 0
                
                # Check if the number is within the valid range of label_list indices
                if number < len(label_list):
                    new_name = f"{label_list[number]} - {column_name}"
                else:
                    new_name = column_name_df  # Retain original if out of bounds
                
                new_column_names.append(new_name)
                matched = True
                break  # Exit after first prefix match
        if not matched:
            new_column_names.append(column_name_df)  # Keep original if no prefix matched
    
    dataframe.columns = new_column_names
    return dataframe
    
```

```python
# Human Labelled
renamded_A1 = rename_columns(A1,column_names,A)
renamded_B1 = rename_columns(B1,column_names,B)
renamded_C1 = rename_columns(C1,column_names,C)
renamded_D1 = rename_columns(D1,column_names,D)


# AI Labelled
renamded_A2 = rename_columns(A2,column_names,A)
renamded_B2 = rename_columns(B2,column_names,B)
renamded_C2 = rename_columns(C2,column_names,C)
renamded_D2 = rename_columns(D2,column_names,D)


human_labelled  = pd.concat([
    renamded_A1[renamded_B1.columns],
    renamded_B1,
    renamded_C1[renamded_B1.columns],
    renamded_D1[renamded_B1.columns]
], ignore_index=True)

# Combine AI Labelled Data
ai_labelled = pd.concat([
    renamded_A2[renamded_B2.columns],
    renamded_B2,
    renamded_C2[renamded_B2.columns],
    renamded_D2[renamded_B2.columns]
], ignore_index=True)
```
